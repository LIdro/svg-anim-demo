```python
# --- NOTEBOOK DOCTOR ---
import json, os
def fix_notebook_metadata():
    path = "/content/svg-anim-demo/notebook/svg_anim_demo.ipynb"
    if not os.path.exists(path): path = "notebook/svg_anim_demo.ipynb"
    try:
        with open(path, 'r') as f: nb = json.load(f)
        if 'widgets' in nb.get('metadata', {}):
            del nb['metadata']['widgets']
            with open(path, 'w') as f: json.dump(nb, f, indent=1)
            print("Fixed widgets metadata.")
    except Exception as e: print(f"Error: {e}")
fix_notebook_metadata()
```
